```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 1. Import the dataset using Pandas from above mentioned url.
data=pd.read_csv("https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
data
```

# 2. High Level Data Understanding:


```python
# a. Find no. of rows & columns in the dataset
print("ROWS AND COLUMNS IN DATA IS  : ",data.shape)
print("Rows are ",data.shape[0])
print("Column are ",data.shape[1])
```

```python
#  b. Data types of columns.
print("DATA TYPES ARE",data.dtypes)
```

```python
#  c. Info & describe of data in dataframe.

print(" cIfo & describe of data in dataframe\n")
data.info()
```

```python
data.describe().T
```

# 3. Low Level Data Understanding :

```python
#  a. Find count of unique values in location column.

print(" count of unique values in location column."   ,   data[["location"]].nunique())
```

```python
#  b. Find which continent has maximum frequency using values counts.
print("frequency of each continent : \n"  , data["continent"].value_counts())
```

```python
#  c. Find maximum & mean value in 'total_cases'.

print("MAX value in Total_cases" , data["total_cases"].max())

print("Mean value in Total_cases ",data["total_cases"].mean() )
```

```python
#  d. Find 25%,50% & 75% quartile value in 'total_deaths'.

print("25% 50% 75% quartile value in total_deaths \n"  , data["total_deaths"].quantile([0.25,0.50,0.75]))
```

```python
# e. Find which continent has maximum 'human_development_index'.
print('continent having maximumm human development index')
data[["continent","human_development_index"]].loc[data["human_development_index"].idxmax()]
```

```python
#f. Find which continent has minimum 'gdp_per_capita'
data[["continent","gdp_per_capita"]].loc[data["gdp_per_capita"].idxmin()]
```

```python
# #4. Filter the dataframe with only this columns
# #['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']
# and update the data frame.
data.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
```

# 5. Data Cleaning

```python
#  a. Remove all duplicates observations
data.drop_duplicates()
```

```python
# Find missing values in all columns
data.isna()
```

```python
# Remove all observations where continent column value is missing
data.dropna(subset=["continent"])
```

```python
# d. Fill all missing values with 0
data.fillna(0)
```

# Date time format 

```python
# a. Convert date column in datetime format using pandas.to_datetime
data["date"]=pd.to_datetime(data["date"])
print(data["date"])
```

```python
# Create new column month after extracting month data from date column
data["month"]=data["date"].dt.month
data["month"]
```

# 7. Data Aggregation:

```python
# Find max value in all columns using groupby function on 'continent' columndat
df1=data.groupby(["continent"]).max()
df1

```

```python
df2=df1.reset_index()
df2
```

```python
df_groupby=pd.DataFrame(data)
df_groupby
```

# 8. Feature Engineering :

```python
# Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
data["total_deaths_to_total_cases"]=data["total_deaths"]/data["total_cases"]
data["total_deaths_to_total_cases"]
```
